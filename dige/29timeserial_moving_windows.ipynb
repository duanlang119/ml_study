```python
%matplotlib inline
import matplotlib.pylab
import pandas as pd
import numpy as np
```

```python
df = pd.Series(np.random.randn(600),index=pd.date_range('7/1/2018',freq='D',periods=600))
```

```python
df.head()
```

```python
r = df.rolling(window=10)
r
```

```python
# r.max,r.median,r.std,r.skew,r.sum,r.var, r.mean
print(r.mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,5))
df.plot(style='r--')
df.rolling(window=10).mean().plot(style='b')
```
